### Selection Sort

**Logic:**
1. **Divide conceptually:** The array is split into a "sorted" region (left) and "unsorted" region (right). Initially, the sorted region is empty.
2. **Find minimum:** In each pass, scan the unsorted region to find the index of the **minimum** element.
3. **Swap:** Swap that minimum with the first element of the unsorted region (i.e., the element at the current pass index).
4. **Expand sorted region:** After each swap, the sorted region grows by one; repeat until the whole array is sorted.

**Idea:** "Select" the smallest remaining element and put it in its correct position from the left.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, after pass 1 we put `1` at index 0, then `2` at index 1, and so on.

---

```python
# Selection Sort — in-place
# Time: O(N²)  |  Space: O(1)
def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            # Use < for ascending (select min); use > for descending (select max)
            if nums[j] < nums[min_index]:
                min_index = j
        nums[i], nums[min_index] = nums[min_index], nums[i]
    return nums

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
selection_sort(nums)
```

**Complexity**

| Case    | Time      | Space |
|--------|-----------|--------|
| **Best**   | O(N²) — still scans unsorted region each time | O(1) |
| **Average**| O(N²)     | O(1)  |
| **Worst**  | O(N²)     | O(1)  |

- **Time:** O(N²)  
  Outer loop runs N times. In pass *i*, the inner loop runs (N − *i* − 1) times.<br>Total comparisons ≈ N + (N−1) + … + 1 = N(N−1)/2 ⇒ **O(N²)**. Same for best, average, and worst — the algorithm always does the same comparisons regardless of input order.

- **Space:** O(1)  
  Sorting is **in-place**; only a fixed number of variables (`i`, `j`, `min_index`) are used. No extra array or recursion.

### Bubble Sort

**Logic:**
1. **Adjacent comparisons:** In each pass, compare adjacent pairs from left to right; if they are out of order, swap them.
2. **Largest "bubbles" right:** Each full pass moves the largest unsorted element to its final position at the end (like a bubble rising).
3. **Shrink the window:** After each pass, the rightmost portion is sorted; repeat over the remaining unsorted prefix.
4. **Early exit (optimization):** If no swap happens in a pass, the array is already sorted — stop early.

**Idea:** Repeatedly sweep through the array, swapping adjacent elements until no more swaps are needed.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, after pass 1 the 9 moves to the end; after pass 2 the 8 is in place; and so on.

---

```python
# Bubble Sort — in-place, with early exit when no swaps
# Time: O(N²) worst/avg; O(N) best when already sorted  |  Space: O(1)
def bubble_sort(nums):
    n = len(nums)
    for i in range(n - 1, 0, -1):   # i = last index of unsorted region
        swapped = False
        for j in range(i):
            # Use > for ascending; use < for descending
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swapped = True
        if not swapped:
            break
    return nums

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
bubble_sort(nums)
```

**Complexity (Bubble Sort)**

| Case       | Time      | Space |
|------------|-----------|--------|
| **Best**   | O(N) — one pass, no swaps, early exit | O(1) |
| **Average**| O(N²)     | O(1)  |
| **Worst**  | O(N²) — all passes, max comparisons   | O(1)  |

- **Time:**  
  **Worst/Average:** O(N²) — up to N−1 passes, with ~i comparisons in pass i ⇒ total ≈ N(N−1)/2.  
  **Best:** O(N) — when the array is already sorted, the first pass does N−1 comparisons, no swaps, and we break.

- **Space:** O(1) — in-place; only a few variables (`i`, `j`, `swapped`) are used.

### Insertion Sort

**Logic:**
1. **Sorted prefix:** The array is conceptually split into a "sorted" region (left) and an "unsorted" region (right). Initially, the first element alone is considered sorted.
2. **Take next element:** In each pass, take the first element of the unsorted region (the element at the current index) as the **key**.
3. **Shift and insert:** Shift all sorted elements that are greater than the key one position to the right, then place the key in the correct position within the sorted region.
4. **Expand sorted region:** After each insertion, the sorted region grows by one; repeat until the whole array is sorted.

**Idea:** Build the sorted array one element at a time by "inserting" each new element into its correct position in the already-sorted prefix, like sorting a hand of cards.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, we keep 9 sorted, then insert 4 before 9, then 3 before 4, then 2, then 8 after 2 but before 9, and so on.

---

```python
# Insertion Sort — in-place
# Time: O(N²) worst/avg; O(N) best when nearly sorted  |  Space: O(1)
def insertion_sort(nums):
    n = len(nums)
    for i in range(1, n):
        key = nums[i]
        j = i - 1
        # Use > for ascending (shift larger elements right); use < for descending
        while j >= 0 and nums[j] > key:
            nums[j + 1] = nums[j]
            j -= 1
        nums[j + 1] = key
    return nums

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
insertion_sort(nums)
```

**Complexity**

| Case       | Time      | Space |
|------------|-----------|--------|
| **Best**   | O(N) — array already sorted, each element inserted with no shifts | O(1) |
| **Average**| O(N²)     | O(1)  |
| **Worst**  | O(N²) — array reverse sorted, every insertion shifts the whole prefix | O(1)  |

- **Time:**  
  **Worst/Average:** O(N²) — in pass *i*, up to *i* comparisons and shifts in the worst case ⇒ total ≈ 1 + 2 + … + (N−1) = N(N−1)/2.  
  **Best:** O(N) — when the array is already sorted, each key is greater than all elements to its left, so the inner loop never runs; one comparison per element.

- **Space:** O(1) — in-place; only a few variables (`i`, `j`, `key`) are used. No extra array or recursion.

### Merge Sort

**Logic:**
1. **Divide:** Recursively split the array into two halves until each half has 0 or 1 element (base case: already sorted).
2. **Conquer:** Two single-element (or empty) segments are trivially sorted.
3. **Merge:** Combine the two sorted halves into one sorted segment by repeatedly taking the smaller of the two front elements and writing it into the result.
4. **Copy back:** The merge step uses a temporary buffer (left/right copies); merged values are written back into the original array.

**Idea:** Divide-and-conquer — sort each half recursively, then merge the two sorted halves in linear time.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, split to `[9,4,3,2]` and `[8,7,1,6,5]`, then recurse; eventually merge sorted `[2,3,4,9]` with `[1,5,6,7,8]` to get `[1,2,3,4,5,6,7,8,9]`.

---

```python
# Merge Sort — divide and conquer, uses extra space for merging
# Time: O(N log N)  |  Space: O(N) for temp arrays + O(log N) call stack
def merge_array(nums, l, mid, r):
    """Merge two sorted segments nums[l:mid+1] and nums[mid+1:r+1] into nums[l:r+1]."""
    left = nums[l : mid + 1]
    right = nums[mid + 1 : r + 1]
    i, j, k = 0, 0, l
    while i < len(left) and j < len(right):
        # Use <= for ascending (stable); use >= for descending
        if left[i] <= right[j]:
            nums[k] = left[i]
            i += 1
        else:
            nums[k] = right[j]
            j += 1
        k += 1
    while i < len(left):
        nums[k] = left[i]
        i, k = i + 1, k + 1
    while j < len(right):
        nums[k] = right[j]
        j, k = j + 1, k + 1

def merge_sort(nums, l, r):
    """Sort nums[l..r] inclusive in-place (r is inclusive)."""
    if l < r:
        mid = (l + r) // 2
        merge_sort(nums, l, mid)
        merge_sort(nums, mid + 1, r)
        merge_array(nums, l, mid, r)

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
merge_sort(nums, 0, len(nums) - 1)  # r is inclusive
nums
```

**Complexity (Merge Sort)**

| Case       | Time        | Space |
|------------|-------------|--------|
| **Best**   | O(N log N)  | O(N)  |
| **Average**| O(N log N)  | O(N)  |
| **Worst**  | O(N log N)  | O(N)  |

- **Time:** O(N log N) in all cases.  
  The recursion tree has O(log N) levels (each level halves the segment). At each level we do O(N) work total (merging all segments at that level). So total work is O(N log N). Input order does not change the number of comparisons.

- **Space:** O(N) for the temporary left/right arrays used during merge (at the top level they sum to N; typically analyzed as O(N) auxiliary space). Plus O(log N) for the recursion call stack.
