### Selection Sort

**Logic:**
1. **Divide conceptually:** The array is split into a "sorted" region (left) and "unsorted" region (right). Initially, the sorted region is empty.
2. **Find minimum:** In each pass, scan the unsorted region to find the index of the **minimum** element.
3. **Swap:** Swap that minimum with the first element of the unsorted region (i.e., the element at the current pass index).
4. **Expand sorted region:** After each swap, the sorted region grows by one; repeat until the whole array is sorted.

**Idea:** "Select" the smallest remaining element and put it in its correct position from the left.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, after pass 1 we put `1` at index 0, then `2` at index 1, and so on.

---

```python
# Selection Sort — in-place
# Time: O(N²)  |  Space: O(1)
def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            # Use < for ascending (select min); use > for descending (select max)
            if nums[j] < nums[min_index]:
                min_index = j
        nums[i], nums[min_index] = nums[min_index], nums[i]
    return nums

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
selection_sort(nums)
```

**Complexity**

| Case    | Time      | Space |
|--------|-----------|--------|
| **Best**   | O(N²) — still scans unsorted region each time | O(1) |
| **Average**| O(N²)     | O(1)  |
| **Worst**  | O(N²)     | O(1)  |

- **Time:** O(N²)  
  Outer loop runs N times. In pass *i*, the inner loop runs (N − *i* − 1) times.<br>Total comparisons ≈ N + (N−1) + … + 1 = N(N−1)/2 ⇒ **O(N²)**. Same for best, average, and worst — the algorithm always does the same comparisons regardless of input order.

- **Space:** O(1)  
  Sorting is **in-place**; only a fixed number of variables (`i`, `j`, `min_index`) are used. No extra array or recursion.

### Bubble Sort

**Logic:**
1. **Adjacent comparisons:** In each pass, compare adjacent pairs from left to right; if they are out of order, swap them.
2. **Largest "bubbles" right:** Each full pass moves the largest unsorted element to its final position at the end (like a bubble rising).
3. **Shrink the window:** After each pass, the rightmost portion is sorted; repeat over the remaining unsorted prefix.
4. **Early exit (optimization):** If no swap happens in a pass, the array is already sorted — stop early.

**Idea:** Repeatedly sweep through the array, swapping adjacent elements until no more swaps are needed.

**Example:** For `[9, 4, 3, 2, 8, 7, 1, 6, 5]`, after pass 1 the 9 moves to the end; after pass 2 the 8 is in place; and so on.

---

```python
# Bubble Sort — in-place, with early exit when no swaps
# Time: O(N²) worst/avg; O(N) best when already sorted  |  Space: O(1)
def bubble_sort(nums):
    n = len(nums)
    for i in range(n - 1, 0, -1):   # i = last index of unsorted region
        swapped = False
        for j in range(i):
            # Use > for ascending; use < for descending
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                swapped = True
        if not swapped:
            break
    return nums

nums = [9, 4, 3, 2, 8, 7, 1, 6, 5]
bubble_sort(nums)
```

**Complexity (Bubble Sort)**

| Case       | Time      | Space |
|------------|-----------|--------|
| **Best**   | O(N) — one pass, no swaps, early exit | O(1) |
| **Average**| O(N²)     | O(1)  |
| **Worst**  | O(N²) — all passes, max comparisons   | O(1)  |

- **Time:**  
  **Worst/Average:** O(N²) — up to N−1 passes, with ~i comparisons in pass i ⇒ total ≈ N(N−1)/2.  
  **Best:** O(N) — when the array is already sorted, the first pass does N−1 comparisons, no swaps, and we break.

- **Space:** O(1) — in-place; only a few variables (`i`, `j`, `swapped`) are used.
