### Extraction of Digits

**Logic:**
1. Take **modulus by 10** → gives the last digit (remainder).
2. Take **floor division by 10** → removes the last digit (quotient).
3. Repeat until the number becomes 0.

**Code:**

```python
n = 12345   # example input
num = abs(n)   # handle negative numbers by working with absolute value

while num > 0:
    last_digit = num % 10
    print(last_digit)
    num = num // 10
```

**Example:** For `n = 12345`, digits are printed as: 5 → 4 → 3 → 2 → 1 (right to left).

#### Count the number of Digits in an Integer

````python
# Method I
n = 5873
count = 0
num = abs(n)
while num > 0:
    count += 1
    num = num // 10
print(count)
````

**Complexity (Method I):**

- **Time:** O(log₁₀(N))  
  The loop runs once per digit. The number of digits in N is ⌊log₁₀(N)⌋ + 1, so we do that many divisions. For this kind of digit-by-digit algorithm, the time complexity depends on the **division base**—here it's 10, so we use log₁₀.

- **Space:** O(1)  
  Only a fixed number of variables (`count`, `num`) are used; no extra structures that grow with input size.

````python
# Method II
from math import log10
count = int(log10(abs(n))) + 1
print(count)
````

#### Check if a number is Palindrome or Not

````python
n = 22122
def is_palindrome(n):
    num = n
    rev = 0
    while num > 0:
        last_digit = num % 10
        rev = rev*10 + last_digit
        num = num//10
    return rev==n
    
is_palindrome(n)
````

**Complexity**

- **Time:** O(log₁₀(N))
- **Space:** O(1)

#### Armstrong Number

An **Armstrong number** (also called a *narcissistic number*) is a number that equals the **sum of its digits, each raised to the power of the number of digits**.

- **Formula:** If the number has `d` digits, then it is Armstrong if  
  **number** = digit₁ᵈ + digit₂ᵈ + … + digitₙᵈ

**Examples:**
- **153** (3 digits): 1³ + 5³ + 3³ = 1 + 125 + 27 = **153** ✓  
- **370** (3 digits): 3³ + 7³ + 0³ = 27 + 343 + 0 = **370** ✓  
- **9** (1 digit): 9¹ = **9** ✓  
- **1634** (4 digits): 1⁴ + 6⁴ + 3⁴ + 4⁴ = 1 + 1296 + 81 + 256 = **1634** ✓



````python
n = 1634
def is_armstrongNum(n):
    num = n
    total = 0
    nod = len(str(num))
    while num>0:
        last_digit = num % 10
        total = total + (last_digit)**nod
        num = num//10
    return total==n

is_armstrongNum(n)
````

**Complexity**

- **Time:** O(log₁₀(N))
- **Space:** O(1)
