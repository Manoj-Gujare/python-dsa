### Factorial of a number using recursion

**Logic:**
1. **Base case:** If `n == 0`, return 1 (0! = 1 by definition).
2. **Recursive case:** n! = n × (n−1)!, so return `n * factorial(n - 1)`.
3. Each call reduces `n` until the base case is reached.

**Code:**

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

**Example:** For `n = 5`: 5! = 5 × 4 × 3 × 2 × 1 = **120**.

---

````python
n = 5
def factorial(n):
    if n == 0 :
        return 1
    return n*factorial(n-1)
factorial(n)
````

**Complexity**

- **Time:** O(N)  
  The function makes exactly N recursive calls (for n, n−1, n−2, … down to 0), and each call does O(1) work.

- **Space:** O(N)  
  The call stack has N frames (one per recursive call from n down to 0), so auxiliary space is linear in n.
