### Factorial of a number using recursion

**Logic:**
1. **Base case:** If `n == 0`, return 1 (0! = 1 by definition).
2. **Recursive case:** n! = n × (n−1)!, so return `n * factorial(n - 1)`.
3. Each call reduces `n` until the base case is reached.

**Code:**

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
```

**Example:** For `n = 5`: 5! = 5 × 4 × 3 × 2 × 1 = **120**.

---

````python
n = 5
def factorial(n):
    if n == 0:              # Base case: without it, recursion would never stop (infinite recursion / stack overflow)
        return 1
    return n*factorial(n-1)
factorial(n)
````

**Complexity**

- **Time:** O(N)  
  The function makes exactly N recursive calls (for n, n−1, n−2, … down to 0), and each call does O(1) work.

- **Space:** O(N)  
  The call stack has N frames (one per recursive call from n down to 0), so auxiliary space is linear in n.

### Reverse an Array

**Logic:**
1. **Two-pointer approach:** Swap elements at `left` and `right`, then move `left` forward and `right` backward until they meet.
2. **Recursion:** Same idea—swap ends, then recurse on the inner subarray; base case when `left >= right`.
3. **Built-in options:** `.reverse()` (in-place) or slicing `[::-1]` (creates a new list).

**Methods below:** Recursion, while loop, `.reverse()`, and slicing.

````python
nums = [5,7,3,2,6,1,5,9]
````

````python
# Method 1: Using recursion (two-pointer via recursion)
# Time: O(N)  |  Space: O(N) [call stack]
n = len(nums)
def reverse_array_rec(arr, left, right):
    if left >= right:
        return
    arr[left], arr[right] = arr[right], arr[left]
    reverse_array_rec(arr, left + 1, right - 1)

rev_rec = nums.copy()
reverse_array_rec(rev_rec, 0, n - 1)
print("Recursion:", rev_rec)
````

````python
# Method 2: Using while loop (two-pointer, in-place)
# Time: O(N)  |  Space: O(1)
def reverse_array_loop(arr, left, right):
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1

rev_loop = nums.copy()
reverse_array_loop(rev_loop, 0, n - 1)
print("While loop:", rev_loop)
````

````python
# Method 3: Built-in .reverse() (in-place)
# Time: O(N)  |  Space: O(1)
rev_inplace = nums.copy()
rev_inplace.reverse()
print(".reverse():", rev_inplace)

# Method 4: Slicing [::-1] (creates new list)
# Time: O(N)  |  Space: O(N)
rev_slice = nums[::-1]
print("Slicing [::-1]:", rev_slice)
````

**Complexity (per method)**

| Method | Time | Space |
|--------|------|--------|
| **Recursion** | O(N) — N/2 swaps, N recursive calls | O(N) — call stack depth N/2 |
| **While loop** | O(N) — N/2 swaps | O(1) — only a few variables |
| **.reverse()** | O(N) — in-place reversal | O(1) — in-place |
| **Slicing [::-1]** | O(N) — copies all elements | O(N) — new list of size N |

- **Recursion:** Each call does O(1) work (one swap); there are ⌈N/2⌉ calls, so time O(N). Stack depth is ⌈N/2⌉ ⇒ space O(N).
- **While loop:** Same number of swaps, no extra recursion ⇒ time O(N), space O(1).
- **.reverse():** Standard in-place reversal ⇒ time O(N), space O(1).
- **Slicing:** Builds a new list of length N ⇒ time O(N), space O(N).
