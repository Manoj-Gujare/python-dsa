### Find the Largest Element in an Array

**Logic:**
1. Assume the first element is the largest (or use `-∞` if you allow empty arrays).
2. Traverse the array once; for each element, update the largest if the current element is greater.
3. Return the largest value after the single pass.

**Idea:** One linear scan, tracking the maximum seen so far. No way to know the max without checking every element, so one pass is optimal.

**Example:** For `[55, 32, -97, 99, 3, 67]`, we keep 55 → 55 → 55 → 99 → 99 → 99 → **99**.

**Edge cases:** Empty array (return `None` or raise); single element (return that element).

```python
# Method I — single pass, track max
# Time: O(N)  |  Space: O(1)
nums = [55, 32, -97, 99, 3, 67]

def largest_no(nums):
    if not nums:
        return None
    largest = nums[0]
    for i in range(len(nums)):
        largest = max(largest, nums[i])
    return largest

largest_no(nums)
```

```python
# Method II — built-in (still O(N) under the hood)
# Time: O(N)  |  Space: O(1)
max(nums)
```

**Complexity**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — One pass; each element compared once. No early exit; we must see every element to guarantee the maximum.
- **Space:** O(1) — Only a fixed number of variables (`largest`, `i`); no extra structures that grow with input size.

### Find the Second Largest Element in an Array (Without Sorting)

**Logic:**
1. We need the second-largest value: strictly smaller than the max, and at least as large as every other element (handling duplicates: e.g. `[1, 2, 2]` → second largest is 1).
2. **Sort approach:** Sort and take the second from the end (after skipping duplicates if required).
3. **Two-pass approach:** First pass find the largest; second pass find the largest element that is strictly less than it.
4. **Single-pass approach:** Maintain `largest` and `second_largest`. When we see a value greater than `largest`, demote `largest` to `second_largest` and update `largest`; else if it's between `second_largest` and `largest`, update `second_largest`.

**Idea:** Either use sorting (simple but O(N log N)) or track the top two in one or two passes for O(N) time and O(1) space.

**Example:** For `[55, 32, 97, -55, 45, 32, 88, 21]`, largest = 97, second largest = **88**.

**Edge cases:** Fewer than two distinct elements (e.g. all same, or one element) — return a sentinel (e.g. `None` or `-∞`) or define behavior for duplicates.

```python
# Method I — Sort and take second from end
# Time: O(N log N)  |  Space: O(1) in-place (e.g. list.sort); O(N) if using sorted()
nums = [55, 32, 97, -55, 45, 32, 88, 21]
nums_sorted = sorted(nums)  # don't mutate original if we need it later
second = nums_sorted[-2]    # for distinct second-largest; else skip duplicates from end
second
```

**Complexity (Method I — Sort)**

| Case       | Time       | Space |
|------------|------------|--------|
| **Best**   | O(N log N) | O(N) with `sorted()` / O(1) in-place with `.sort()` |
| **Average**| O(N log N) | same |
| **Worst**  | O(N log N) | same |

- **Time:** O(N log N) — dominated by sorting.
- **Space:** O(N) if using `sorted()` (new list); O(1) auxiliary if using in-place `.sort()` (stack for recursion typically not counted).

```python
# Method II — Two passes: first find largest, then find max element strictly less than it
# Time: O(N)  |  Space: O(1)
def second_largest_two_pass(nums):
    largest = float("-inf")
    s_largest = float("-inf")
    for x in nums:
        largest = max(largest, x)
    for x in nums:
        if s_largest < x < largest:
            s_largest = x
    return s_largest if s_largest != float("-inf") else None

second_largest_two_pass(nums)
```

**Complexity (Method II — Two passes)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — two linear passes (or one for max + one for second).
- **Space:** O(1) — only a few variables.

```python
# Method III — Single pass (optimal): maintain largest and second_largest as we go
# Time: O(N)  |  Space: O(1)
def second_largest_one_pass(nums):
    largest = float("-inf")
    s_largest = float("-inf")
    for x in nums:
        if x > largest:
            s_largest = largest
            largest = x
        elif s_largest < x < largest:
            s_largest = x
    return s_largest if s_largest != float("-inf") else None

second_largest_one_pass(nums)
```

**Complexity (Method III — Single pass)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — one pass; each element is compared at most twice (once against `largest`, once against `s_largest`). Constant work per element ⇒ O(N).
- **Space:** O(1) — only `largest` and `s_largest` (and loop variable); no extra structures.

### Check if the Array Is Sorted

**Logic:**
1. For **ascending** order: every adjacent pair must satisfy `arr[i] ≤ arr[i+1]`. If any pair has `arr[i] > arr[i+1]`, the array is not sorted.
2. Traverse from index `0` to `n-2`; compare each element with the next. If we never find a violation, the array is sorted.
3. For **strictly ascending** (no duplicates), use `<` instead of `≤` when defining "sorted" if required.

**Idea:** One linear pass; the first violation tells us it's not sorted. We need to check every pair to confirm sorted.

**Example:** `[3, 5, 6, 4, 8, 9, 10, 20]` — at index 2 we have 6 > 4 ⇒ **not sorted**. `[1, 2, 3, 4]` — no violation ⇒ **sorted**.

**Edge cases:** Empty or single-element array is trivially sorted. All equal elements is sorted for non-strict order.

```python
# Single pass: first violation ⇒ not sorted
# Time: O(N)  |  Space: O(1)
def is_sorted_asc(nums):
    if not nums or len(nums) == 1:
        return True
    for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:
            return False
    return True

nums = [3, 5, 6, 4, 8, 9, 10, 20]
is_sorted_asc(nums)
```

```python
# Variant: one-liner (built-in) — creates a copy for comparison
# Time: O(N log N) from sort  |  Space: O(N)
nums == sorted(nums)
```

**Complexity**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(1) — first pair violates (e.g. `[2, 1, 3, 4, ...]`) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) — fully sorted, check all n−1 pairs | O(1) |

- **Time:** Best O(1) when we find a violation early; worst O(N) when the array is sorted or the only violation is at the end. Average case O(N).
- **Space:** O(1) — only loop index and comparisons; no extra storage.

### Remove Duplicates from a Sorted Array

**Logic:**
1. Array is **sorted**, so duplicate values appear consecutively. We need the first occurrence of each value at the front and the **length** of that unique prefix (elements beyond can be ignored).
2. **Two-pointer (optimal):** Use a write index `i`; the first element is always kept. Scan with `j` from 1 to end. When `nums[j] != nums[i]`, we've found a new unique value — increment `i`, copy `nums[j]` to `nums[i]`. Return `i + 1` as the count of unique elements.
3. **Using extra space:** Iterate and collect unique values (e.g. in a dict/set that preserves order); then write them back. Simpler but uses O(k) space for k unique elements.

**Idea:** Because the array is sorted, a single forward pass is enough. One pointer tracks "where to write the next unique"; the other scans. No need for extra array if we overwrite in place.

**Example:** `[1, 1, 1, 2, 3, 4, 4, 7, 9, 9, 9, 10]` → unique prefix `[1, 2, 3, 4, 7, 9, 10]`, length **7**. Trailing elements can stay as-is; only the first 7 positions matter.

**Edge cases:** Empty array → return 0. Single element → return 1. All same → return 1.

```python
# Method - Using dict to collect unique (order preserved in Python 3.7+)
# Time: O(2N) ~ O(N) |  Space: O(k) for k unique elements
def remove_duplicates_dict(nums):
    if not nums:
        return 0
    seen = {}
    for x in nums:
        seen[x] = None  # or use dict.fromkeys(nums)
    uniq = list(seen.keys())
    for i in range(len(uniq)):
        nums[i] = uniq[i]
    return len(uniq)

nums2 = [1, 1, 1, 2, 3, 4, 4, 7, 9, 9, 9, 10]
k2 = remove_duplicates_dict(nums2)
print("Length:", k2)
print("Unique prefix:", nums2[:k2])
```

**Complexity (Method — Dict / extra space)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(k) |
| **Average**| O(N) | O(k) |
| **Worst**  | O(N) | O(k) |

- **Time:** O(N) — one pass to build the dict, one pass to write back; k = number of distinct elements.
- **Space:** O(k) — the dict (and the list of keys) stores up to k unique values. Not O(1) in-place.

---

```python
# Method — Two-pointer in-place (optimal)
# Time: O(N)  |  Space: O(1)
def remove_duplicates_sorted(nums):
    if not nums:
        return 0
    n = len(nums)
    i = 0  # write index; nums[0..i] is the unique prefix
    for j in range(1, n):
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]
    return i + 1

nums = [1, 1, 1, 2, 3, 4, 4, 7, 9, 9, 9, 10]
k = remove_duplicates_sorted(nums)
print("Length:", k)
print("Unique prefix:", nums[:k])
```

**Complexity (Method Two-pointer in-place)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — single pass; each element is compared and at most written once.
- **Space:** O(1) — only the two indices `i` and `j`; no extra array or hash structure.

### Right Rotate an Array by K Places

**Logic:**
1. **Right rotate by k** means the last k elements move to the front: `[..., last k]` become the first k, and the first n−k elements shift right. So result = **last k elements** + **first n−k elements**.
2. Use **k = k % n** so that k > n (e.g. k = 10, n = 6) or k = 0 are handled; after that, 0 ≤ k < n.
3. **Brute:** Repeat k times: pop from end, insert at front. Correct but slow when k is large (each insert at 0 is O(n)).
4. **Slicing:** Assign `nums[:] = nums[n-k:] + nums[:n-k]`. One assignment; creates temporary lists so O(N) space.
5. **Three reverses (optimal in-place):** View array as [A][B] where B = last k, A = first n−k. We want [B][A]. Reverse B, reverse A, then reverse the whole → (B′A′)′ = BA. So: reverse(last k), reverse(first n−k), reverse(entire array). O(N) time, O(1) space.

**Idea:** Right by k = take the last k and put them in front. Either do it with slices (simple, O(N) space) or with three in-place reverses (O(1) space).

**Example:** `[3, 9, 5, 6, 7, 2]`, k = 3 → last 3 are `[6, 7, 2]`, first 3 are `[3, 9, 5]` → result **`[6, 7, 2, 3, 9, 5]`**.

**Edge cases:** Empty array (do nothing). k = 0 or k = n (no change). k > n → use k % n.

```python
# Method I — Brute: repeat k times — pop from end, insert at front
# Time: O(k × N)  |  Space: O(1)
def right_rotate_brute(nums, k):
    if not nums:
        return
    n = len(nums)
    k = k % n
    for _ in range(k):
        e = nums.pop()
        nums.insert(0, e)

nums = [3, 9, 5, 6, 7, 2]
right_rotate_brute(nums, 3)
nums
```

**Complexity (Method I — Pop/insert)**

| Case       | Time      | Space |
|------------|-----------|--------|
| **Best**   | O(N) — when k = 0 or k = n (no iterations) | O(1) |
| **Average**| O(k × N)  | O(1) |
| **Worst**  | O(N²) — when k = n−1, we do n−1 inserts each O(n) | O(1) |

- **Time:** Each `insert(0, e)` shifts the whole list ⇒ O(n) per rotation; k rotations ⇒ O(k × n). Worst when k = Θ(n) ⇒ O(n²).
- **Space:** O(1) — only a few variables; list modified in place (aside from list growth/shrink).

```python
# Method II — Slicing: last k + first n−k (use k % n)
# Time: O(N)  |  Space: O(N) — slicing creates new lists
def right_rotate_slice(nums, k):
    if not nums:
        return
    n = len(nums)
    k = k % n
    if k == 0:
        return
    nums[:] = nums[n - k:] + nums[:n - k]

nums = [3, 9, 5, 6, 7, 2, 10, 9]
right_rotate_slice(nums, 5)
nums
```

**Complexity (Method II — Slicing)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(N) |
| **Average**| O(N) | O(N) |
| **Worst**  | O(N) | O(N) |

- **Time:** O(N) — building the two slices and the concatenated list is linear.
- **Space:** O(N) — `nums[n-k:]` and `nums[:n-k]` create temporary lists; the RHS is a new list of size n before assignment.

```python
# Method III — Three reverses (optimal in-place)
# Time: O(N)  |  Space: O(1)
def reverse(nums, left, right):
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1

def right_rotate_reverse(nums, k):
    if not nums:
        return
    n = len(nums)
    k = k % n
    if k == 0:
        return
    # [A][B] → reverse B, reverse A → B'A' → reverse all → BA
    reverse(nums, n - k, n - 1)
    reverse(nums, 0, n - k - 1)
    reverse(nums, 0, n - 1)

nums = [3, 9, 5, 6, 7, 2, 10, 9]
right_rotate_reverse(nums, 5)
nums
```

**Complexity (Method III — Three reverses)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — each reverse touches a contiguous segment once; total elements touched is O(n).
- **Space:** O(1) — only the `reverse` helper and indices; no extra array. Truly in-place.

### Move Zeros to the End of the List

**Logic:**
1. We need to move all zeros to the end while keeping the relative order of non-zero elements unchanged.
2. **Brute:** Build a temp list of non-zeros, then overwrite the original: first fill with non-zeros, then fill the rest with zeros. Correct but uses O(N) extra space.
3. **Optimal (two-pointer):** Use index `i` as the "write" position for the next non-zero, and `j` to scan. When `nums[j] != 0`, swap `nums[i]` with `nums[j]` and advance `i`. After one pass, all non-zeros are at the front and zeros have been pushed right. O(N) time, O(1) space.

**Idea:** Either copy non-zeros to a new structure then write back (simple, O(N) space), or use a single pass with two pointers and swap non-zeros into place so zeros naturally end up at the end.

**Example:** `[1, 0, 2, 4, 3, 0, 0, 3, 5, 1]` → **`[1, 2, 4, 3, 3, 5, 1, 0, 0, 0]`**. Relative order of 1, 2, 4, 3, 3, 5, 1 is preserved.

**Edge cases:** All zeros (no change in layout). No zeros (one pass, no swaps). Single element (return as-is).

```python
# Method I — Brute: temp list of non-zeros, then overwrite
# Time: O(N)  |  Space: O(N)
nums = [1, 0, 2, 4, 3, 0, 0, 3, 5, 1]
n = len(nums)
temp = [x for x in nums if x != 0]
nz = len(temp)
for i in range(nz):
    nums[i] = temp[i]
for i in range(nz, n):
    nums[i] = 0
print(nums)
```

**Complexity (Method I — Brute)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(N) |
| **Average**| O(N) | O(N) |
| **Worst**  | O(N) | O(N) |

- **Time:** O(N) — one pass to collect non-zeros, one pass to write them back, one pass to fill zeros.
- **Space:** O(N) — temp list holds all non-zero elements in the worst case (no zeros).

```python
# Method II — Optimal: two-pointer, swap non-zeros into place (in-place)
# Time: O(N)  |  Space: O(1)
nums = [1, 0, 2, 4, 3, 0, 0, 3, 5, 1]
n = len(nums)
i = 0
while i < n and nums[i] != 0:
    i += 1
if i == n:
    print(nums)  # no zeros
else:
    j = i + 1
    while j < n:
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
        j += 1
    print(nums)
```

**Complexity (Method II — Two-pointer in-place)**

| Case       | Time | Space |
|------------|------|--------|
| **Best**   | O(N) | O(1) |
| **Average**| O(N) | O(1) |
| **Worst**  | O(N) | O(1) |

- **Time:** O(N) — single pass with `j`; each element is read once and at most swapped once.
- **Space:** O(1) — only two indices `i` and `j`; no extra array. Truly in-place.

```python
# Same two-pointer idea, compact form (e.g. for LeetCode 283)
def moveZeroes(nums):
    """Modify nums in-place: move all 0's to the end, preserve order of non-zeros."""
    i = 0
    for j in range(len(nums)):
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
```
